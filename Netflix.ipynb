```python
for col in df:
  if df[col].dtype=='int' or df[col].dtype=='float':
    print(col.capitalize())
    print(f"Mean:{df[col].mean()}")
    print(f"Median:{df[col].median()}")
    print(f"Minimum of {col}:{df[col].min()}")
    print(f"Maximum of {col}:{df[col].max()}")
    print(f"Variance of {col}:{df[col].var()}")
    print(f"Standard deviation:{df[col].std()}")
    print()
    plt.figure(figsize=(8,5))
    plt.boxplot(df[col]) # for detecting outliers, uses numerical values
    plt.title(col)
    plt.show()
    print("_________________________________________________________________________________________________________________")
    print()
```

```python
import matplotlib.pyplot as plt

# Sorting and selecting the top 10 movies by rating
top_10_movies = df.sort_values(by='rating', ascending=False).head(10)

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(top_10_movies['title'], top_10_movies['rating'], marker='o', color='blue', linestyle='-')

plt.xlabel('Movie Title', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.title('Top 10 Movies Based on Rating', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

```python
import matplotlib.pyplot as plt

# Group by 'type' and 'Age_Suitability' and count the occurrences
age_suitability_count = df[['type', 'Age_Suitability']].groupby('type').value_counts().unstack().fillna(0)

# Plotting the histogram
age_suitability_count.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Age Suitability Distribution by Type', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Age Suitability', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Group by 'type' and 'Age_Suitability' and count the occurrences
age_suitability_count = df[['type', 'Age_Suitability']].groupby('type').value_counts().unstack().fillna(0)

# Plotting the histogram (stacked bar plot)
age_suitability_count.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Age Suitability Distribution by Type', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Age Suitability', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plotting the count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', hue='type', data=df)

plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Count of Movies by Rating and Type', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Type', loc='upper right')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Filter movies from the United States
us_movies = df[df['country'] == 'United States']

# Get the top 5 most frequent genres in the United States
us_genre_count = us_movies['genres'].value_counts().head(5)

# Plotting the pie chart using plt.pie
plt.figure(figsize=(8, 8))
plt.pie(us_genre_count, labels=us_genre_count.index, autopct='%1.1f%%', startangle=90,
        wedgeprops={'edgecolor': 'black'})

# Title and layout
plt.title('Top 5 Movie Genres in the United States', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Get the value counts for 'type'
type_counts = df['type'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(7, 7))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Distribution of Content Type', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
